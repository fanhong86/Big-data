```python
rate_file=sc.textFile("ratings.csv")
rate_file.take(4) 
```

```python
rateheader = rate_file.take(1)[0]
rate_data = rate_file.filter(lambda line: line!=rateheader).map(lambda line: line.split(",")).map(lambda x: (int(x[0]),int(x[1]),float(x[2]))).cache()


```

```python
rate_data.take(10)
```

```python
movie_file=sc.textFile("movies.csv")
movie_file.take(4) 
```

```python
movieheader = movie_file.take(1)[0]
movie_data = movie_file.filter(lambda line: line!=movieheader).map(lambda line: line.split(",")).map(lambda x: (int(x[0]),str(x[1]),str(x[2]))).cache()

```

```python
movie_data.take(10)
```

```python
def get_genre(id):
    genrelist=movie_data.filter(lambda x: x[0]==id).map(lambda x:x[2])
    return genrelist.collect()
```

```python
testuser=1
testmovie=1172
avg=3
```

```python
get_genre(testmovie)
```

```python
rate_users = rate_data.filter(lambda x: x[1]==testmovie)
```

```python
rate_users.collect()
```

```python
like_users = rate_users.filter(lambda x:x[2]>avg).map(lambda x:x[0]).collect()
```

```python
type(like_users)
```

```python
co_movies=rate_data.filter(lambda x: x[0] in like_users).filter(lambda x: x[2] > avg)
```

```python
co_movies.collect()
```

```python
movie_keys = co_movies.map(lambda x: (x[1],x[2]))
```

```python
movie_keys.collect()
```

```python
total_rating = movie_keys.mapValues(lambda x: (x, 1)).reduceByKey(lambda x, y: (x[0] + y[0], x[1] + y[1]))
total_rating.collect()
```

```python
avg_rating=total_rating.mapValues(lambda x:x[0]/x[1])
```

```python
avg_rating.collect()
```

```python
recommendation=avg_rating.sortBy(lambda x:x[1],ascending=False)
```

```python
recommendation.collect()
recommendation.take(5)
```
